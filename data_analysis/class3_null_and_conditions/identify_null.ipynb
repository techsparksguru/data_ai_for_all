```python
import pandas as pd
df = pd.read_csv('../height_weight.csv')
```

```python
# Observe the count of values in each column
df.info()
```

```python
# Are there any null/NaN values in the dataframe
df.isnull()
```

```python
# Which columns have at least one null value ?
df.isnull().any()
```

```python
# Print and see 
df[['height','weight','repwt']].head(20)
```

```python
# Which columns have all null values ?
df.isnull().all()
```

```python
# Which rows have at least one null value?
df.isnull().any(axis=1)
```

```python
#which rows have ALL null values ?
df.isnull().all(axis=1)
```

```python
# Counting is difficult, so chain any() again. This tells us there is every row with at least one value
df.isnull().all(axis=1).any()
```

```python
# Get all rows where at least one value in the row is null
df[df.isnull().any(axis=1)]
```

```python
# Which columns do NOT have null value ?
df.notnull().any()
```

```python
df['repwt'].head(20)
```

```python
# Get rows that have MIN_TEMP_GROUND null
df[df['repwt'].isnull()]
```

```python
# Get rows that have MIN_TEMP_GROUND not null
df[df['repwt'].notnull()]
```

```python
list_of_not_null_indices = df[df['repwt'].notnull()].index.values.tolist()
```

```python
# Define a range starting from 5 until 200 with steps of 2, convert to list
x = list(range(5,200,2))
for i in x:
    print(i)
```

```python
if x == list_of_not_null_indices:
    print("equal")
```

```python
# Drop all rows that have MIN_TEMP_GROUND null 
df.drop(df[df['repwt'].isnull()].index.values.tolist())
```
