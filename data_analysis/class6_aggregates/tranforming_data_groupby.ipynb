```python
import pandas as pd
athletes = pd.read_csv('../height_weight.csv')
athletes.dropna(inplace=True)
```

```python
athletes.info()
```

# GROUP BY

```python
# See total wight and height by each sex 
athletes.groupby(['sex']).sum()
```

```python
# What if I need to sum only on height and NOT on all
athletes.groupby(['sex'])[['height']].sum()
```

```python
# Get mean weight and height for each sex
athletes.groupby(['sex'])[['weight','height']].mean()
```

```python
# Lets add a new column to the dataframe - sport which the athlete belongs to
import random
import numpy as np
athletes['sport'] = np.nan
for i in range(0,athletes.count()[0]):
    athletes.loc[athletes.index[i],'sport'] = random.choice(['cricket','football','aquatics','badminton','basketball','soccer','table tennis'])
athletes
```

```python
# Get mean weight and height for each sport category, and then by sex as well
athletes.groupby(['sport','sex'])[['weight','height']].mean()
```

```python
# Get mean weight and height for each sport category, and then by sex as well and rename columns
athletes.groupby(['sport','sex'])[['weight','height']].mean().rename(columns={"weight":"avg_weight","height":"avg_height"})
```
