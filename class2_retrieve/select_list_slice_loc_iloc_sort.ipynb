```python
import pandas as pd
df = pd.read_csv('../height_weight.csv')
df
```

```python
# Index with column
df['height']
```

```python
df['height'][1]
```

```python
# Using python list (array) to select multiple values
df[['height', 'weight', 'sex']]
```

```python
# slice always selects rows. Try df[:4], df[0:]
df[2:4][['height','weight']]
```

```python
# Transpose
dft=df.T
# slice on strings includes the last item, whereas integers dont
dft['weight':'height']
```

```python
# loc and iloc only for row based indexing
# loc indexes only by label (except....see below)
dft.loc['weight':'height']
```

```python
# TIME to PRESSURE rows, and 0,1,2 columns
dft.loc['weight':'height',[0,1,2]]
```

```python
#iloc is only for location based indexing
dft.iloc[[0,3],[1,2]]
```

```python
# Column name is not printed like this
dft.iloc[:,2]
```

```python
# 0th row
df.loc[[0]]
```

```python
# 0th row and height column
df.loc[[0],['height']]
```

```python
# Get all rows that have TEMP above mean TEMP
df[df['height'] > df['weight'].mean()]
```

```python
# Sort rows
df.sort_index()
```

```python
# Sort columns
df.sort_index(axis=1)
```

```python
# Observe that the original dataframe still is NOT sorted
df
```

```python
# use inplace=True to modify the frame
df.sort_index(axis=1,inplace=True)
```

```python
df.sort_index(axis=1, ascending=False)
```
