```python
import pandas as pd
```

```python
df = pd.read_csv('Real_Estate_Sales_2001-2017.csv')
```

```python
df.info()
```

```python
df
```

# Homes greater than average home price in Andover

```python
df[(df['SaleAmount'] > df['SaleAmount'].mean()) & (df['Town'] == 'Andover')]
```

# Average home price by Property Type

```python
pd.set_option('display.float_format', lambda x: '%.1f' % x)
df.groupby(['PropertyType'])['SaleAmount'].mean()
```

```python
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.html
%matplotlib inline
df.groupby(['PropertyType'])['SaleAmount'].mean().plot()
```

# Is assessed value correct ?

```python
# Positive correlation between AssessedValue and SaleAmount
df['AssessedValue'].corr(df['SaleAmount'])
```

```python
# Mean sale amount
df['SaleAmount'].mean()
```

# How many people paid between [range]

```python
# How many paid between my budget [300,000,350,000]
df[df['SaleAmount'].between(300000,350000)]
```

```python
# How many people paid as per my budget [300000,350000] grouped by Town
my_df = df[df['SaleAmount'].between(300000,350000)]
my_df
```

# How many people paid within my budget, grouped by Town

```python
# How many people paid as per my budget [300000,350000] grouped by Town
my_df1 = my_df.groupby(['Town']).agg(COUNT=("ID","count"))
my_df1
```

# Which city has most folks who have same budget as mine

```python
# Which town has most number of people that have paid the same as my budget?
my_df1[my_df1['COUNT'] == my_df1['COUNT'].max()]
```

```python
my_df1.sort_values(by=['COUNT'], ascending=False)
```
